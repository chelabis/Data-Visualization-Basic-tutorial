### DataFrame Plot

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0), 'B': np.random.randn(365).cumsum(0) + 20,
                   'C': np.random.randn(365).cumsum(0) - 20}, index = pd.date_range('1/1/2024', periods = 365))
df.head()
```

```python
df.plot();
```

```python
df.plot('A', 'B', kind = 'scatter');    # Same as df.plot.scatter('A', 'B')
```

```python
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis').set_aspect('equal');
```

```python
df.plot.box()
```

```python
df.plot.hist(alpha=0.7, bins=20);
```

```python
df.plot.kde();     # Kernel Density Estimate plot
```

### pandas.tools.plotting

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns = iris.feature_names)
df['Species'] = y
df.head()
```

```python
df.Species.replace(0, 'setosa', inplace=True)
df.Species.replace(1, 'versicolor', inplace=True)
df.Species.replace(2, 'virginica', inplace=True)
```

```python
df.head()
```

```python
pd.plotting.scatter_matrix(df);
```

```python
plt.figure()
pd.plotting.parallel_coordinates(df, 'Species');
```
