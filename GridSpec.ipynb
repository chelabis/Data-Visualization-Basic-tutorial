## Histograms

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0, len(axs)):
    size = 10**(n+1)
    sample = np.random.normal(loc=0, scale=1, size=size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title(f'n={size}')
```

```python
Y = np.random.normal(loc=0, scale=1, size=10000)
X = np.random.random(10000)

plt.scatter(X,Y);
```

```python
import matplotlib.gridspec as gs
```

```python
plt.figure(figsize=(5,5))
gspec = gs.GridSpec(3,3)

top_hist = plt.subplot(gspec[0, 1:])
side_hist = plt.subplot(gspec[1:, 0])
right = plt.subplot(gspec[1:, 1:])

right.scatter(X, Y)
top_hist.hist(X, bins=100, density=True, color='r', alpha=0.5)
side_hist.hist(Y, bins=100, density=True, orientation='horizontal', color='g', alpha=0.5);

side_hist.invert_xaxis()
```

### Box

```python
normal_smaple = np.random.normal(loc=0, scale=1, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_smaple, 'random': random_sample, 'gamma': gamma_sample})
df.head()
```

```python
df.describe()
```

```python
plt.boxplot([df.normal, df.random, df.gamma], whis=[0,100]);
```

```python
plt.figure(figsize=(5,5))

plt.boxplot([df.normal, df.random, df.gamma], whis=[0,100])
ax2 = plt.gca().inset_axes([0, 0.6, 0.6, 0.4])
ax2.hist(df.gamma, bins=100, density=True)
ax2.yaxis.tick_right();
```
