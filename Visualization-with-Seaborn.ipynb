```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60, 15, 1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50, 150, 5), label='v1')
plt.hist(v2, alpha=0.7, bins=np.arange(-50, 150, 5), label='v2')
plt.legend();
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3);
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

### Joint plot

```python
sns.jointplot(v1, alpha=0.4);
```

```python
grid = sns.jointplot(v1, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(v1, kind='hex');
```

```python
sns.set_style('white')

sns.jointplot(v1, kind='kde', space=0);
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns = iris.feature_names)
df['Species'] = y
df.Species.replace(0, 'setosa', inplace=True)
df.Species.replace(1, 'versicolor', inplace=True)
df.Species.replace(2, 'virginica', inplace=True)
df.head()
```

```python
sns.pairplot(df, hue='Species', diag_kind='kde');
```
